```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.svm import SVC

data=pd.read_csv('breast.csv')

x=data.iloc[:,1:-1].values
y=data.iloc[:,-1].values

# data techniques 
# missing data

# from sklearn.impute import SimpleImputer 

# imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
# imputer.fit(x[// which location])

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.linear_model import LogisticRegression


log=SVC(C=1)
log.fit(x_train,y_train)

y_pred=log.predict(x_test)

from sklearn.metrics import confusion_matrix 

confuse=confusion_matrix(y_pred,y_test)

print(confuse)

from sklearn.metrics import accuracy_score
aoc=accuracy_score(y_pred,y_test)

aoc

# k fold cross validation

from sklearn.model_selection import cross_val_score

accuracies = cross_val_score(estimator=log, X=x_train, y=y_train, cv=10)

# Calculate mean accuracy
mean_accuracy = accuracies.mean() * 100

# Calculate standard deviation of accuracies
std_accuracy = accuracies.std() * 100

# Print mean accuracy and standard deviation
print("Accuracy: {:.2f}%".format(mean_accuracy))
print("Standard deviation of accuracy: {:.2f}%".format(std_accuracy))






```
